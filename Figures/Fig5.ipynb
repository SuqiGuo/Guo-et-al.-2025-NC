```python
SCENARIO_COMBNATION = ['crop-ctl','frst-ctl','irri-crop']
region_list= ['NAM', 'CE', 'CSAM', 'CAF', 'ESAM', 'SAF', 'SEAS']
labels = [
    "WNA",  # Western North America
    "EES",  # East Europe Siberia
    "WAB",  # Western Amazon Basin
    "CCB",  # Central Congo Basin
    "EAB",  # Eastern Amazon Basin
    "SCP",  # Southern Congo Periphery
    "EC"    # Eastern China
]
scenario_title=['CROP-CTL','FRST-CTL','IRR-CROP']
y_axis_text1=[' Eluc per 100 million\nhectares of land (GtC)','','']
y_axis_text2=['Eluc per 100 million hectares\n of cropland expension (GtC)','Eluc per 100 million hectares\n    of afforestation (GtC)','  Eluc per 100 mm of\n annual irrigation (GtC)']
number=['a','b','c','d','e','f']
```

```python
# this script calculate the BGC effects perarea of LUC. for example, the BGC effects of unit area of afforestation or cropland expansion. this can only conduct for CESM and MPI-ESM.
fig = plt.figure(constrained_layout= False)
#fig.subplots_adjust(wspace= 10, hspace= 0)
subfigs = fig.subfigures(2, 3,  wspace= 0, hspace= 0)
#subfigs.layout(None)
#subfigs.subplots_adjust(wspace=0, hspace=0)
#fig.tight_layout(pad= 0)
fig.set_size_inches(15, 6)
for j, SCENARIO in enumerate(SCENARIO_COMBNATION):
 mpi_value = []
 cesm_value = []
 ecearth_value = []
 mpi_perLCCLM = []
 cesm_perLCCLM = []
 for i, region_source in enumerate(region_list):     
    mpiesm_cland_dir_cLand = xr.open_dataset('/work/bm1147/b380949/web-monitoring/MPI-PLOT/Lutdata/TCRE/perarea/' + region_source + \
                                             '/Eluc_' + SCENARIO + '_mpiesm_Lut_last30mean_chname_' + region_source + '.nc')
     # mpiesm_cland_dir_cLand.cLand_local.values[0][0][0]*-1000: the direct AFOLU emission(regional integrated local BGC effects) of $region_source;
     # mpiesm_cland_dir_cLand.cLand_nonlocal.values[0][0][0]*-1000:the regional integrated nonlocal BGC effect of $region_sources(caused by the global AFOLU induced nonlocal BGP effects on this unchanged region-$region_source)     
    cesm_cland_dir_cLand= xr.open_dataset('/work/bm1147/b380949/web-monitoring/CESM/plot/final_Lunit/RTCRE_regional/perarea/' + region_source + \
                                          '/Eluc_' + SCENARIO + '_cesm_last30mean_' + region_source + '.nc')
    ecearth_cland_dir_cLand = xr.open_dataset('/work/bm1147/b380949/web-monitoring/ecearth_plot/RTCRE_regional/perarea/' + region_source + \
                                              '/Eluc_' + SCENARIO + '_ecearth_last30mean_gridlevelsep_' + region_source + '.nc')
    mpiesm_luc = xr.open_dataset('/work/bm1147/b380949/LUC_models/mpiesm/regional/' + SCENARIO +'_' + region_source + '_mpiesm_fldmean_chname.nc')
    cesm_luc = xr.open_dataset('/work/bm1147/b380949/LUC_models/cesm/regional/' + SCENARIO +'_' + region_source + '_cesm_fldmean_chname.nc')
    # if region_source == 'NAM':
    mpi=mpiesm_cland_dir_cLand.cLand_local.values[0][0][0]*1000  # 1 EgC=1000 GtC multiplying 1000 is used to converting EgC to GtC
    cesm=cesm_cland_dir_cLand.cLand_local.values[0][0][0]*1000
    ecearth=ecearth_cland_dir_cLand.cLand_local.values[0][0][0]*1000
    #mpiesm_luc.total_interpolated =the cover fraction of cropland expension   
    if j==1:
        ecearth=np.nan
    else:
        ecearth=ecearth_cland_dir_cLand.cLand_local.values[0][0][0]*1000
        
    if j==2:
        mpidata=float(mpiesm_cland_dir_cLand.cLand_local.values[0][0][0]*1000/mpiesm_luc.total_interpolated)*100    
        cesmdata=float(cesm_cland_dir_cLand.cLand_local.values[0][0][0]*1000/cesm_luc.total_interpolated)*100
    else:
        mpidata=float(mpiesm_cland_dir_cLand.cLand_local.values[0][0][0]*1000/mpiesm_luc.total_interpolated)    
        cesmdata=float(cesm_cland_dir_cLand.cLand_local.values[0][0][0]*1000/cesm_luc.total_interpolated)
    ecearth_value.append(ecearth)
    cesm_value.append(cesm)
    mpi_value.append(mpi)
    mpi_perLCCLM.append(mpidata)
    cesm_perLCCLM.append(cesmdata)
     
 ensemble_arr=[cesm_value,mpi_value,ecearth_value]
 ensemble_arr2=[cesm_perLCCLM,mpi_perLCCLM] 
 ensemble= np.nanmean(ensemble_arr,axis=0)
 ensemble2= np.nanmean(ensemble_arr2,axis=0)
    
 subfig = subfigs[0,j]
 ax = subfig.subplots(1, 1)
 plt.figure(figsize=(17.5, 6))
 #labels = ['CAF','CE','CSAM','ESAM','NAM','SAF','SEAS']
 labels_1 = ['','','','','','','']
 x = [1,1.25,1.5,1.75,2,2.25,2.5]  # the label locations
    
 ax.set_position([0.016, 0.01, 0.8, 0.93])
  
 rects1 = ax.scatter(x, mpi_value,marker= '^',s=100, color="#8C2D04", label='mpiesm')
 rects2 = ax.scatter(x, cesm_value,marker= 'o',s=100, color="#8C2D04", label='cesm') 
 rects3 = ax.scatter(x, ecearth_value,marker= 'X',s=100, color="#8C2D04", label='ecearth')
 rects10 = ax.scatter(x, ensemble, marker= '*', s=500, color="#8C2D04", label='mean')
 ax.set_xticks(x)
 ax.tick_params(axis='x', direction='in', length=4, labelsize=12,labelbottom=False)
 ax.tick_params(axis='y', direction='out', length=4, labelsize=12)
 ax.text(0.37, 1.04, scenario_title[j], fontsize=14, rotation=0,fontweight='bold',
     verticalalignment='center', horizontalalignment='left', transform=ax.transAxes)
 ax.text(-0.21, 0.5,y_axis_text1[j], fontsize=14, rotation=90,fontweight='bold',
         verticalalignment='center', horizontalalignment='left', transform=ax.transAxes)   
 ax.text(0.01, 1.06,number[j] , fontsize=14, rotation=0,fontweight='bold',
         verticalalignment='top', horizontalalignment='left', transform=ax.transAxes)  
 ax.yaxis.set_major_locator(MaxNLocator(nbins=6))
 subfig = subfigs[1,j]
 ax2 = subfig.subplots(1, 1)
 plt.figure(figsize=(17.5, 6))
 ax2.set_position([0.016, 0.01, 0.8, 0.93])
 rects1n = ax2.scatter(x, mpi_perLCCLM,marker= '^',s=100, color="#8C2D04", label='mpiesm')
 rects2n = ax2.scatter(x, cesm_perLCCLM,marker= 'o',s=100, color="#8C2D04", label='cesm')   
 rects11 = ax2.scatter(x, ensemble2, marker= '*', s=500, color="#8C2D04", label='mean')  

 ax2.set_xticks(x)
 ax2.tick_params(axis='x', direction='in', length=4, labelsize=12)
 ax2.tick_params(axis='y', direction='out', length=4, labelsize=12)
 ax2.set_xticklabels(labels, fontsize=13)
 ax2.text(-0.21, 0.5,y_axis_text2[j], fontsize=14, rotation=90,fontweight='bold',
         verticalalignment='center', horizontalalignment='left', transform=ax2.transAxes)
 ax2.text(0.008, 1.06,number[3+j] , fontsize=14, rotation=0,fontweight='bold',
         verticalalignment='top', horizontalalignment='left', transform=ax2.transAxes)  
 ax2.yaxis.set_major_locator(MaxNLocator(nbins=6))
    
 all_values_ax1 = np.concatenate([mpi_value,cesm_value,ecearth_value,ensemble])
 all_values_ax1 = all_values_ax1[np.isfinite(all_values_ax1)]
 #ax1.set_ylim([np.min(all_values_ax1) -5, np.max(all_values_ax1) +5])
 ax.set_ylim([np.min(all_values_ax1) -3, np.max(all_values_ax1) +3])
 all_values_ax2 = np.concatenate([mpi_perLCCLM,cesm_perLCCLM,ensemble2])
 all_values_ax2 = all_values_ax2[np.isfinite(all_values_ax2)]
 if j==2:
     ax2.set_ylim([np.min(all_values_ax2) *1.4, np.max(all_values_ax2)*1.4])
 else:
     ax2.set_ylim([np.min(all_values_ax2) -3, np.max(all_values_ax2) +3])

 ax.axhline(0,color='grey',linewidth = 0.7)
 ax2.axhline(0,color='grey',linewidth = 0.7)
    #ax1.legend([rects1,rects3,rects5,rects10], ['mpiesm','cesm','ecearth','ensemble mean'], fontsize = 20) 
    #rects1.refline(y = 0, color = "grey", linewidth = 0.0001)#2ca25f
 legend_elements = [Line2D([0], [0], marker='^', color='w', label='MPI-ESM ',markerfacecolor='#A0A0A0', markersize=13),
                   Line2D([0], [0], marker='o', color='w', label='CESM ', markerfacecolor='#A0A0A0', markersize=12),
                   Line2D([0], [0], marker='X', color='w', label='EC-Earth ',markerfacecolor='#A0A0A0', markersize=13),
                   Line2D([0], [0], marker='*', color='w', label='Ensemble mean ', markerfacecolor='#A0A0A0', markersize=26),
                   Patch(color='#8C2D04', label='Local BGC'),
                   ] 
font_properties = FontProperties()
font_properties.set_family('serif')  # Set font family
font_properties.set_name('Times New Roman')  # Specific font name
#font_properties.set_weight('bold')  # Set font weight to bold
font_properties.set_size(13)  # Set font size

fig.legend(handles=legend_elements, loc='lower center',bbox_to_anchor=(0.45, -0.17),ncol=5,prop=font_properties )                   
#fig.savefig('/work/bm1147/b380949/web-monitoring/secP/supplement_cland_regional_compare/NC-Fig5_Eluc_perarea_regional_compare.jpg', format = 'jpg', bbox_inches='tight', pad_inches=0.1, dpi=300)
    #ax1.legend([(rects1,rects2),(rects3,rects4),(rects5,rects6)], ['mpiesm','cesm','ecearth'], fontsize = 20)
fig.savefig('/scratch/b/b380949/' + 'NC-Fig5_Eluc_perarea_regional_compare.pdf', format = 'pdf', bbox_inches='tight', pad_inches=0.1, dpi=300)
```
